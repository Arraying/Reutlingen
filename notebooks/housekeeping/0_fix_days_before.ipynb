# Fix days_before_forecast_day
We assume that when you are running this pipeline, the inference pipeline has already been run for every day before today. Therefore, we can set the days_before_forecast_day to 1 for all the days before today.

```python
from datetime import datetime, timedelta

import hopsworks
import os

import pandas as pd
```

```python
if os.environ.get("HOPSWORKS_API_KEY") is None:
    os.environ["HOPSWORKS_API_KEY"] = open("../.hw_key").read()

project = hopsworks.login()
fs = project.get_feature_store()
today = datetime.now() - timedelta(0)
```

```python
monitor_fg = fs.get_feature_group("aq_predictions")
monitor_df = monitor_fg.filter(monitor_fg.date <= today).read()
monitor_df
```

```python
# For all the days before today, we need to set the days_before_forecast_days to 1
monitor_fg = fs.get_feature_group("aq_predictions")
monitor_df = monitor_fg.filter(monitor_fg.date <= today).read()
monitor_df["days_before_forecast_day"] = 1
monitor_df
```

```python
monitor_fg.insert(monitor_df, write_options={"wait_for_job": True})
```
